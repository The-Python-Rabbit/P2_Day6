```python
# ------------------------------------------------------------------------------------------
# Product Inventory Project with Unit-Testing - 5 July 2021
#         - Includes Extre Challenge (update and delate a product).
#         - Includes Extra Challenge 2 (test any class or function), In this case, the first instance of a Book class.
#-------------------------------------------------------------------------------------------

import doctest
doctest.testmod()

class Product:
    
    """Defines the Product Class from which the Book and Movie classes are derived"""
    
    def __init__(self, cat, id_code, title, price, quantity):
        self.__cat = cat
        self.__id_code = id_code
        self.__title = title
        self.__price = float(price)
        self.__quantity = int(quantity)
        
        
    def print_info(self):
        total = self.__price* self.__quantity
        
#
#        Assign some of the atributes to short variables for readability purposes
        si = self.__id_code
        st = self.__title
        sq = str(self.__quantity)

        return(si + "  " + st.ljust(36) + f"${self.__price:,.2f}".rjust(6," ") + sq.rjust(12," ") + f"${total:,.2f}".rjust(13," "))
    
    def get_total(self):
        return self.__price* self.__quantity
    
    def get_cat(self):
        return self.__cat
    
    def get_code(self):
        return self.__id_code
    
    def update_price(self, price):
        self.__price = price
        
    def update_quantity(self, quantity):
        self.__quantity = quantity
        
    def file_out(self):
        return(f"{self.__cat},{self.__id_code},{self.__title},{self.__price},{self.__quantity}")       
        
        
class Book(Product):
    
#    In this test, I try to emulate the process by which I entered data in the initial part of the main program by
#    creating a dummy list and a dummy string. The test parses the string to extract the various class attributes
#    and then creates a new class instance in the first row of the dummy list.
#    I thought that this would make for a more thorough test then simply placing values directly in the initialization
#    statement of the class.


    """Book class with an author information and cat attribute of 'BOOK'. Test for first instance of Book class.
 
    >>> temp_list = []
    >>> temp = ("BOOK","PROD001","The Rainbow Fish","J. Alison James",12.99,30)
    >>> temp_list.append(Book(temp[0], temp[1], temp[2], temp[3], float(temp[4]), int(temp[5])))
    >>> temp_list[0].print_info()
    'PROD001  The Rainbow Fish                    $12.99          30      $389.70'

    """

    category = "BOOK"
    
    def __init__(self, cat, id_code, title, author, price, quantity):
        super().__init__(cat, id_code, title, price, quantity)
        self.__author = author
        
    def book_file_out(self):
        return(f"{super().file_out()},{self.__author}")
      
class Movie(Product):
    """Movie class with a genre and release year private attributes and a cat attribute of 'MOVIE'"""
    
    category = "MOVIE"
    
    def __init__(self, cat, id_code, title, genre, year, price, quantity):
        super().__init__(cat, id_code, title, price, quantity)
        self.__genre = genre
        self.__year = year
       
    def movie_file_out(self):
        return(f"{super().file_out()},{self.__genre},{self.__year}")
        
# -------------------------------------------------------------------------------------------------------------
#  Main program
#----------------------------------------------------------------------------------------------------------------

# Read in .csv file and parse to determine if it is a Book or a Movie. Then store class instances in a list of classes.

infile = open("productlist.csv")
prod_list = []

for info in infile:
    temp = info.rsplit(",")
    if temp[0].strip() == "BOOK":
         prod_list.append(Book(temp[0].strip(), temp[1].strip(), temp[2].strip(), temp[3].strip(), float(temp[4]), int(temp[5])))
    else:
        prod_list.append(Movie(temp[0].strip(), temp[1].strip(), temp[2].strip(), temp[3].strip(), temp[4].strip(), float(temp[5]), int(temp[6])))     
    
# *** Main Menu ***
    
print("\n*** PRODUCT INVENTORY PROGRAM: ***\n")
print("1. Show All Products")
print("2. Add a Product")
print("3. Update a Product")
print("4. Delete a Product")
print("5. Exit")

#-------------------------------------------------------------------------------------
# Choice 1: Show all of the products
#-------------------------------------------------------------------------------------

choice = 1

# Begin "choice" while loop
while choice:
    try:
        choice = int(input("\nEnter your choice: "))
    except:
        print ("\nImproper entry. Please try again")
        continue
        
    if choice == 1:
        e1 = 0
        sum_total = 0
        blank = " "
        print("\nPRODUCT  NAME",blank * 31,"PRICE", blank * 2,"QUANTITY", blank * 6, "TOTAL")
        print("="*80)
        for emp in prod_list:
            print (prod_list[e1].print_info())
            sum_total += prod_list[e1].get_total()
            e1 +=1
        print("-"*80)

        print(f"TOTAL:", f"${sum_total:,.2f}".rjust(69, " "))
        
#----------------------------------------------------------------------------------------
# Choice 2: Enter a new Product, either Book or Movie
#----------------------------------------------------------------------------------------
        
    elif choice == 2:
        t_cat = input("Enter new product category ((B)OOK, (M)OVIE):")
        if t_cat[0].lower() == "b":
            t_code = input("Enter Product Code: ").upper()
            t_title = input("Enter Title: ")
            t_author = input("Enter Author: ")
            t_price = float(input("Enter price: "))
            t_quan = int(input("Enter quantity: "))
            prod_list.append(Book("BOOK", t_code, t_title, t_author, t_price, t_quan))  
            
        elif t_cat[0].lower() == "m":
            t_code = input("Enter Product Code: ").upper()
            t_title = input("Enter Title: ")
            t_genre = input("Enter Genre: ")
            t_year = input("Enter Year: ")
            t_price = float(input("Enter price: "))
            t_quan = int(input("Enter quantity: "))
            prod_list.append(Movie("MOVIE", t_code, t_title, t_genre, t_year, t_price, t_quan))
            
        else:
            print ("\nImproper entry. Please try again.")

#--------------------------------------------------------------------------
# Choice 3: Update Product Information, either price or quantity
#--------------------------------------------------------------------------
            
    elif choice == 3:
        u_match = False
        
# Begin while loop:        
        while u_match == False:
            t_code = input("Enter Product Code to Update: ").upper()
            i1 = 0
            
#
# Check if user has entered valid Product Code. Return True if Product Code is valid (i.e., exists)

            for prod in prod_list:
                if t_code == prod_list[i1].get_code():
                    u_match = True
                    continue
                else:
                    i1 +=1
            if u_match == False:
                print("Invalid Product Code: Please try again.")
                continue
                
# End while loop               
#
# Update price and/or quantity if user entered valid Product Code

            print("\nUpdate Menu:")
            print("1. Update Price")
            print("2. Update Quantity")
            print("3. Finished with Updates")
            u_choice = 1
            while u_choice:
                u_choice = int(input("\nPlease enter update option: "))
                if u_choice == 1:
                    t_price = float(input("Enter new price: "))
                    prod_list[i1].update_price(t_price)
                elif u_choice == 2:
                    t_quan = int(input("Enter new quantity: "))
                    prod_list[i1].update_quantity(t_quan)
                elif u_choice == 3:
                    break
                else:
                    print("\nImproper entry. Please try again.")                   
 
#------------------------------------------------------------------------------
# Choice 4: Delete a Product.
#------------------------------------------------------------------------------
    elif choice == 4:
        u_match = False
        
# Begin while loop:
        while u_match == False:
            t_code = input("Enter Product Code to Delete: ").upper()
            i1 = 0
            
#
# Check if user has entered valid Product Code. This is the same procedure as used for Update Product information segment.

            for prod in prod_list:
                if t_code == prod_list[i1].get_code():
                    u_match = True
                    continue
                else:
                    i1 +=1
            if u_match == False:
                print("Invalid Product Code: Please try again.")
                continue
# End while loop  

#
# Delete object
            del prod_list[i1]
            print("Product deleted")

#-------------------------------------------------------------------------
# Choice 5: Exit progam
#-------------------------------------------------------------------------

    elif choice == 5:
        break
        
#-------------------------------------------------------------------------
# Error: Prompt user for new entry if he/she entered improper choice.
#-------------------------------------------------------------------------

    else:
        print("\nImproper entry. Please try again.")
        
# End "choice" while loop
   
# Create file for output. Overwrite if file exists.
#

fileout = open("product_final.csv","w")
i1 = 0
for prod in prod_list:
    temp = prod_list[i1].get_cat()
    if temp == "BOOK":
        fileout.write(f"{prod_list[i1].book_file_out()}\n")
    else:
        fileout.write(f"{prod_list[i1].movie_file_out()}\n")
    i1+=1
    
fileout.close()
print("Information printed to product_final.csv")
print("*** Thank you for your patronage ***")
```
